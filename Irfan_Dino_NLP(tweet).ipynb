```python
import pandas as pd
data = pd.read_csv('/content/sample_data/tweets.csv')
data.head()
```

```python
list = ['id',	'keyword',	'location']
data1 = data.drop((list), axis=1)
data1
```

```python
data1.info()
```

```python
category = pd.get_dummies(data1['target']).rename(columns=lambda x:'star_'+str(x))
new_data = pd.concat([data1, category], axis=1)
new_data = new_data.drop(columns='target')
new_data
```

```python
new_dat = new_data.rename({'text':'tweet','star_0': 'no_disaster',	'star_1':'disaster'}, axis=1)
new_dat
```

```python
from nltk.stem.api import StemmerI
import re
import string
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

def remove_URL(tweet):
    url = re.compile(r'https?://\S+|www\.\S+')
    return url.sub(r'', tweet)

def remove_punct(tweet):
    translator = str.maketrans('', '', string.punctuation)
    return tweet.translate(translator)

stop = set(stopwords.words('english'))
def remove_stopwords(tweet):
    filtered_words = [word.lower() for word in tweet.split() if word.lower() not in stop]
    return ' '.join(filtered_words)

stemmer =  PorterStemmer()
def remove_stem(tweet):
    stemmered_word = [stemmer.stem(word) for word in tweet.split()]
    return ' '.join(stemmered_word)

lemmatizer = WordNetLemmatizer()
def remove_lemmatizer(tweet):
    lemmatizered_word = [lemmatizer.lemmatize(word) for word in tweet.split()]
    return ' '.join(lemmatizered_word)
```

```python
new_dat['tweet'] = new_dat.tweet.map(remove_URL)
new_dat['tweet'] = new_dat.tweet.map(remove_punct)
new_dat['tweet'] = new_dat.tweet.map(remove_stopwords)
new_dat['tweet'] = new_dat.tweet.map(remove_stem)
new_dat['tweet'] = new_dat.tweet.map(remove_lemmatizer)
```

```python
new_dat.tweet
```

```python
from collections import Counter
def counter_word(tweet_col):
    count = Counter()
    for tweet in tweet_col.values:
        for word in tweet.split():
            count[word] += 1
    return count

counter = counter_word(new_dat.tweet)
```

```python
num_unique = len(counter)
```

```python
text = new_dat['tweet']
target = new_dat[['no_disaster', 'disaster']]
```

```python
from sklearn.model_selection import train_test_split
train_text, val_text, train_target, val_target = train_test_split(text, target, test_size=0.2)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

token = Tokenizer(num_words=num_unique, oov_token='x')
token.fit_on_texts(train_text)
token.fit_on_texts(val_text)

train_seq = token.texts_to_sequences(train_text)
val_seq = token.texts_to_sequences(val_text)

train_pad = pad_sequences(train_seq)
val_pad = pad_sequences(val_seq)
```

```python
model = tf.keras.Sequential([
        tf.keras.layers.Embedding(input_dim=num_unique, output_dim=16),
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, dropout=0.2)),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(2, activation='softmax')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy')>0.9):
      print("\nAkurasi telah mencapai >90%!")
      self.model.stop_training = True
callbacks = myCallback()

num_epochs=20
history = model.fit(train_pad, train_target, epochs=num_epochs,
                    validation_data=(val_pad, val_target), verbose=2)
```
